```python
def ratio_error(v1,e1,v2,e2):
  #v1, e1: numerator value and error
  #v2, e2: denominator value and error  
  error1 = e1/v2
  error2 = (e2/v2)*(v1/v2)
  error = np.sqrt(error1*error1+error2*error2)
  return error
```

## Jet-$R_{\mathrm{AA}}$

```python
# Prepare arrays of Jet-pT
jet_pt_pp_in = jets_pp[:,1]
jet_pt_pbpb_in = jets_pbpb[:,1]

jet_pt_pp1_in = jets_pp1[:,1]
jet_pt_pbpb1_in = jets_pbpb1[:,1]

jet_pt_pp2_in = jets_pp2[:,1]
jet_pt_pbpb2_in = jets_pbpb2[:,1]

jet_pt_pp3_in = jets_pp3[:,1]
jet_pt_pbpb3_in = jets_pbpb3[:,1]

jet_pt_pp4_in = jets_pp4[:,1]
jet_pt_pbpb4_in = jets_pbpb4[:,1]

jet_pt_pp5_in = jets_pp5[:,1]
jet_pt_pbpb5_in = jets_pbpb5[:,1]


bins_11 = [200,250,300,400,500,1000]
pt_min = 200
pt_max = 1000

# Fill Histogram
n_pp, pt = np.histogram(jet_pt_pp_in, bins= bins_11 )
n_pbpb, pt = np.histogram(jet_pt_pbpb_in, bins= bins_11 )

n_pp1, pt = np.histogram(jet_pt_pp1_in, bins= bins_11 )
n_pbpb1, pt = np.histogram(jet_pt_pbpb1_in, bins= bins_11 )

n_pp2, pt = np.histogram(jet_pt_pp2_in, bins= bins_11 )
n_pbpb2, pt = np.histogram(jet_pt_pbpb2_in, bins= bins_11 )

n_pp3, pt = np.histogram(jet_pt_pp3_in, bins= bins_11 )
n_pbpb3, pt = np.histogram(jet_pt_pbpb3_in, bins= bins_11 )

n_pp4, pt = np.histogram(jet_pt_pp4_in, bins= bins_11 )
n_pbpb4, pt = np.histogram(jet_pt_pbpb4_in, bins= bins_11 )

n_pp5, pt = np.histogram(jet_pt_pp5_in, bins= bins_11 )
n_pbpb5, pt = np.histogram(jet_pt_pbpb5_in, bins= bins_11 )

# Statistical Errors
err_n_pp = np.sqrt(n_pp)
err_n_pbpb = np.sqrt(n_pbpb)

err_n_pp1 = np.sqrt(n_pp1)
err_n_pbpb1 = np.sqrt(n_pbpb1)

err_n_pp2 = np.sqrt(n_pp2)
err_n_pbpb2 = np.sqrt(n_pbpb2)

err_n_pp3 = np.sqrt(n_pp3)
err_n_pbpb3 = np.sqrt(n_pbpb3)

err_n_pp4 = np.sqrt(n_pp4)
err_n_pbpb4 = np.sqrt(n_pbpb4)

err_n_pp5 = np.sqrt(n_pp5)
err_n_pbpb5 = np.sqrt(n_pbpb5)

# bin width
dpt = (pt[1:]-pt[:-1])
# bin center
pt = pt[0:-1] + 0.5*dpt


n_ev_pp = 315769
n_ev_pp1 = 338726 
n_ev_pp2 = 352683 
n_ev_pp3 = 369925
n_ev_pp4 = 380962 
n_ev_pp5 = 389140

n_ev_pbpb = 6461
sigma = 5.15
d_eta = 2*2

# Jet Spectrum
dn_dpt_pp = (n_pp*sigma)/n_ev_pp/dpt/d_eta
dn_dpt_pbpb = (n_pbpb*sigma)/n_ev_pbpb/dpt/d_eta

dn_dpt_pp1 = (n_pp1*sigma)/n_ev_pp1/dpt/d_eta
dn_dpt_pbpb1 = (n_pbpb1*sigma)/n_ev_pbpb/dpt/d_eta

dn_dpt_pp2 = (n_pp2*sigma)/n_ev_pp2/dpt/d_eta
dn_dpt_pbpb2 = (n_pbpb2*sigma)/n_ev_pbpb/dpt/d_eta

dn_dpt_pp3 = (n_pp3*sigma)/n_ev_pp3/dpt/d_eta
dn_dpt_pbpb3 = (n_pbpb3*sigma)/n_ev_pbpb/dpt/d_eta

dn_dpt_pp4 = (n_pp4*sigma)/n_ev_pp4/dpt/d_eta
dn_dpt_pbpb4 = (n_pbpb4*sigma)/n_ev_pbpb/dpt/d_eta

dn_dpt_pp5 = (n_pp5*sigma)/n_ev_pp5/dpt/d_eta
dn_dpt_pbpb5 = (n_pbpb5*sigma)/n_ev_pbpb/dpt/d_eta

# Errors 
err_dn_dpt_pp = (err_n_pp*sigma)/n_ev_pp/dpt/d_eta
err_dn_dpt_pbpb = (err_n_pbpb*sigma)/n_ev_pbpb/dpt/d_eta

err_dn_dpt_pp1 = (err_n_pp1*sigma)/n_ev_pp1/dpt/d_eta
err_dn_dpt_pbpb1 = (err_n_pbpb1*sigma)/n_ev_pbpb/dpt/d_eta

err_dn_dpt_pp2 = (err_n_pp2*sigma)/n_ev_pp2/dpt/d_eta
err_dn_dpt_pbpb2 = (err_n_pbpb2*sigma)/n_ev_pbpb/dpt/d_eta

err_dn_dpt_pp3 = (err_n_pp3*sigma)/n_ev_pp3/dpt/d_eta
err_dn_dpt_pbpb3 = (err_n_pbpb3*sigma)/n_ev_pbpb/dpt/d_eta

err_dn_dpt_pp4 = (err_n_pp4*sigma)/n_ev_pp4/dpt/d_eta
err_dn_dpt_pbpb4 = (err_n_pbpb4*sigma)/n_ev_pbpb/dpt/d_eta

err_dn_dpt_pp5 = (err_n_pp5*sigma)/n_ev_pp5/dpt/d_eta
err_dn_dpt_pbpb5 = (err_n_pbpb5*sigma)/n_ev_pbpb/dpt/d_eta

# Generate Plots
fig = plt.figure()

# Calculate RAA and error
raa =  dn_dpt_pbpb/dn_dpt_pp
error_raa = ratio_error(dn_dpt_pbpb,err_dn_dpt_pbpb,dn_dpt_pp,err_dn_dpt_pp)

raa1 =  dn_dpt_pbpb1/dn_dpt_pp1
error_raa1 = ratio_error(dn_dpt_pbpb1,err_dn_dpt_pbpb1,dn_dpt_pp1,err_dn_dpt_pp1)

raa2 =  dn_dpt_pbpb2/dn_dpt_pp2
error_raa2 = ratio_error(dn_dpt_pbpb2,err_dn_dpt_pbpb2,dn_dpt_pp2,err_dn_dpt_pp2)

raa3 =  dn_dpt_pbpb3/dn_dpt_pp3
error_raa3 = ratio_error(dn_dpt_pbpb3,err_dn_dpt_pbpb3,dn_dpt_pp3,err_dn_dpt_pp3)

raa4 =  dn_dpt_pbpb4/dn_dpt_pp4
error_raa4 = ratio_error(dn_dpt_pbpb4,err_dn_dpt_pbpb4,dn_dpt_pp4,err_dn_dpt_pp4)

raa5 =  dn_dpt_pbpb5/dn_dpt_pp5
error_raa5 = ratio_error(dn_dpt_pbpb5,err_dn_dpt_pbpb5,dn_dpt_pp5,err_dn_dpt_pp5)

err_1 = ratio_error(raa1,error_raa1,raa,error_raa)

err_2 = ratio_error(raa2,error_raa2,raa,error_raa)

err_3 = ratio_error(raa3,error_raa3,raa,error_raa)

err_4 = ratio_error(raa4,error_raa4,raa,error_raa)

err_5 = ratio_error(raa5,error_raa5,raa,error_raa)


##########

plt.errorbar(pt, raa2/raa, ms=9,mfc='red',mec='black',fmt='o', label=r"R=0.4",
             xerr=0.5*dpt, yerr=err_2, color='red')

plt.errorbar(pt, raa3/raa, ms=9,mfc='blue',mec='black',fmt='o', label=r"R=0.6",
             xerr=0.5*dpt, yerr=err_3, color='blue')

plt.errorbar(pt, raa4/raa, ms=9,mfc='green',mec='black',fmt='o', label=r"R=0.8",
             xerr=0.5*dpt, yerr=err_4, color='green')

plt.errorbar(pt, raa5/raa, ms=9,mfc='cyan',mec='black',fmt='o', label=r"R=1.0",
             xerr=0.5*dpt, yerr=err_5, color='cyan')



#axes setting
plt.legend(loc='upper left')
plt.xscale('log')
plt.xlabel(r"$p^{\mathrm{jet}}_{\mathrm{T}}$ [GeV]")
plt.ylabel(r"$R^{\mathrm{R}}_{\mathrm{AA}}/R^{\mathrm{R=0.2}}_{\mathrm{AA}}$")
plt.xlim(pt_min,pt_max)
plt.ylim(0.8,2.0)
plt.axhline(1, color = "blue", linestyle="dashed", linewidth=1.5) 
#plt.text(pt_min+5,-0.05, r"PbPb(0-10%)"r"$\sqrt{s_{\mathrm{NN}}}=5.02 TeV$")
#plt.text(pt_min+5,-0.15,r'$anti-k_{\mathrm{t}},R=0.4,(|y_{\mathrm{jet}}|<2)$')
#plt.text(pt_min+5,-0.25,r"$JS(MATTER+LBT),Q_{\mathrm{sw}}=2GeV$")

# save plot to the Desktop
plt.tight_layout()
plt.savefig('cms_double_pT_0-10')
print("R=0.3/0.2=", raa1/raa)
print("R=0.4/0.2=",raa2/raa)
print("R=0.6/0.2=",raa3/raa)
print("R=0.8/0.2=",raa4/raa)
print("R=1.0/0.2=",raa5/raa)
```
